```
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 2*np.pi, 50)
x = np.sin(t)
y = np.cos(t)

# Now create a figure
plt.figure()
# and plot x inside it
plt.plot(x)

# Now create a new figure
plt.figure()
# and plot y inside it...
plt.plot(y)
# Add a title
plt.title("Cos")
```

```
x = np.linspace(0, 2*np.pi, 50)
y1 = np.sin(x)
y2 = np.sin(2*x)
plt.figure() # Create figure
plt.plot(y1)
plt.plot(x, y1)

# red dot-dash circle
plt.plot(x, y1, 'r')

# red marker only circle
plt.plot(x, y1, 'r-o')

# clear figure then plot 2 curves
plt.clf()
plt.plot(x, y1, 'g-o', x, y2, 'b-+')
plt.legend(['sin(x)','sin(2x)'])
```

```
import numpy as np

# Simple array creation
a = np.array([0, 1, 2, 3, 4, 5])
print(a)
print(type(a))
print(a.dtype)
print(a.ndim)
print(a.shape)
print(a.itemsize)
print(a.nbytes)
```

```
# Multi-dimensional array

a = np.array([[ 0, 1, 2, 3],
              [10,11,12,13]])
print(a)
print(a.shape)
print(a.size)
print(a.ndim)

# Get and set elements in Multi-dimensional array
print(a[1, 3])
a[1, 3] = -1
print(a[1, 3])
print(a[1])

# Slicing array
print(a)

# print all row and 3rd column
print(a[:,2])

# print first row and all column 
print(a[0,:])

# negative indices work also
print(a)
print(a[:,:-1])
```

Create a 2D array of 5 by 5 and find the min and max values from each row and each column

```
import numpy as np
x = np.random.rand(5,5)*10
print(x,'\n')
a = np.amax(x, 0)
b = np.amin(x, 0)
c = np.amax(x, 1)
d = np.amin(x, 1)
print('Max and Min Value Column 0: ')
print(a[0] , b[0] )
print('Max and Min Value Column 1: ')
print(a[1] , b[1] )
print('Max and Min Value Column 2: ')
print(a[2] , b[2] )
print('Max and Min Value Column 3: ')
print(a[3] , b[3] )
print('Max and Min Value Column 4: ')
print(a[4] , b[4],'\n')
print('Max and Min Value Row 0: ')
print(c[0] , d[0] )
print('Max and Min Value Row 1: ')
print(c[1] , d[1] )
print('Max and Min Value Row 2: ')
print(c[2] , d[2] )
print('Max and Min Value Row 3: ')
print(c[3] , d[3] )
print('Max and Min Value Row 4: ')
print(c[4] , d[4] )
```

Create an array of size 100 and compute the mean, median, standard deviation

```
import numpy as np
x = np.random.rand(10,10)

print('Mean: ')
print(np.mean(x),'\n')
print('Median: ')
print(np.median(x),'\n')
print('Standard Deviation')
print(np.std(x),'\n')
```

Normalize a 5x5 random matrix
